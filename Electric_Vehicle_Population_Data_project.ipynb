```python
#Handling outliers
R1=df.quantile(0.25)
R2=df.quantile(0.75)
IQR=R2-R1
print(IQR)
```

```python
#Formula for Handling outliers
df=df[~((df<(R1-1.5*IQR))|(df>(R2+1.5*IQR))).any(axis=1)]
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

ENCODING

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
Re=LabelEncoder()
x["Vechicle ID"]=Re.fit_transform(x["Vechicle ID"])
x["City"]=Re.fit_transform(x["City"])
x["EcarBrand"]=Re.fit_transform(x["EcarBrand"])
x["Car Model"]=Re.fit_transform(x["Car Model"])
x["Electric Vehicle Type"]=Re.fit_transform(x["Electric Vehicle Type"])
x
```

```python
df.head()
```

```python
x.head()
```

```python
# splitting data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#importing the model
from sklearn.linear_model import LinearRegression
```

```python
#step2 -: initalize the model
model=LinearRegression()
model
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
from sklearn.metrics import r2_score
print(f"Accuracy : {r2_score(ytest, ypred)}")
```

```python
print(mean_absolute_error(ytest,ypred))
```

```python
print(mean_squared_error(ytest,ypred))
```

```python
print(r2_score(ytest,ypred))
```

```python
print(model.score(xtrain,ytrain))
```
